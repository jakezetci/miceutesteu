# Метод сращиваемых асимптотических разложений на простом примере


##  Метод сращиваемых ассимптотических разложений - общая схема


Необходимио найти решение краевой задачи для следующего дифференциального уравнения
\begin{equation} \label{eq_L}
L(y, x; \varepsilon) = 0, 
\end{equation}
и малых значения параметра  $\varepsilon$.

Здесь $L$ - дифференциальный оператор, x принадлежит области $\Omega \subset R^n $, $y(x; \varepsilon)$ - искомая функция.

Для данного уравнения должны быть поставленя корректные граничные условия на $\delta \Omega$ границе области $\Omega$.


Пусть решение (\ref{eq_L}) при $\varepsilon \rightarrow 0$  найти в явном виде (или численно) не удается. Будем искать решение 
разлагая функцию $y(x; \varepsilon)$  в ряд по $\varepsilon$.
\begin{equation}
y(x; \varepsilon) = \sum \limits_{k=0}^{\infty} \varepsilon^k y_k(x).
\end{equation}

Обозначим $n$-частичную сумму $\tilde{y}_n (x, \varepsilon) = \sum \limits_{i=0}^{n} \varepsilon^i y_i(x)$.
Ряд $\sum \limits_{i=0}^{\infty} \varepsilon^i y_i(x)$ является  формальным ассимптотическим решением дифференциального 
уравнения $L(y, x; \varepsilon) = 0$ (или краевой задачи, если разложение удовлетворяет граничным условиям) если

\begin{equation}\label{uslovie_shodimosti}
| L(\tilde{y}_n (x, \varepsilon), x; \varepsilon) | < M \varepsilon^n
\end{equation}

Если последнее соотношение и граничные условия выполняются для любых $x \in \Omega \cup \partial \Omega$, то задача регулярная и нам не интересна.  Если в части области $\Omega \cup \delta \Omega$ соотношение не выполняется, то говорят, что задача является {\it сингулярно возмущенной}. Мы будем рассматривать задачи в которых условие (\ref{uslovie_shodimosti}) не выполняется только в области $\Gamma$ на части границы  $\partial \Omega$ и ее окрестности. Эта область называется пограничным слоем.

В случае задач с пограничными слоями ряд $\sum \limits_{i=0}^{\infty} \varepsilon^i y_i(x)$ называется внешним разложением.
Функции $y_i(x)$ - это решения вспомогательных задач, которые обычно легче, чем основная задача. Уравнения для $y_i(x)$ получаются подстановкой ряда в исходное дифференциальное уравнение.  
Для регулярной части границы (не входящей в погранслой) граничные условия для $y_i(x)$ получить просто. Для части границы входяшей в погранслой граничные условия можно поставить только зная поведение функции в погранслое. То есть надо решать внутренюю задачу.

Для решения задачи в пограничном слое вводят растянутые координаты (называемые внутренними переменными) $X = X(x,\varepsilon)$ (oбычно формула простая, $X = x/\varepsilon^{\alpha}$, $\alpha >0$) и ищут внутреннее асимпотическое разложение, которое хорошо описывает функцию $y(x,\varepsilon)$ в области $\Gamma$ (погранслое):
\begin{equation}
y(x; \varepsilon) = Y(X; \varepsilon))= \sum \limits_{i=0}^{\infty} \mu_i(\varepsilon) Y_i(X).
\end{equation}
($\mu_i(\varepsilon)$ - калибровочная последовательность.)

Задачи определения области погранслоя $\Gamma$ и калибровочной последовательности $\mu_i(\varepsilon)$ не всегда являются тривиальными. Если мы с ними справились, то уравнения для $Y_i(X)$  легко выписываются, если  ряд $\sum \limits_{i=0}^{\infty} \mu_i(\varepsilon) Y_i(X)$ подставить в исходное уравнение $L(y, x; \varepsilon) = 0$.

Уравнения для функций $y_i (x)$ внешнего разложения и $Y_i(X)$ внутреннего разложения найти несложно, но, вообще говоря, определить эти функции однозначно нельзя (обычно, но бывают исключения), так как из краевых условий изначальной задачи нельзя однозначно определить краевые условия для разложений. Откуда же взять недостающие условия, достаточные для однозначного определения функций $u_k (x)$ и $U_i(X)$?

Заметим, что ряды внешнего и внутреннего разложения должны приближать одну и ту же функцию $u(x, \varepsilon)$, но в разных областях. Положим, что есть достаточно широкая (по сравнению с погранслоем) область, где эти ряды одновременно приближают решение и поэтому ассимпотически (при $\varepsilon \rightarrow 0$) cовпадают.  Именно существование области, где внутреннее и внешнее разложение асимптотически совпадают и позволяет однозначно определить $u_k (x)$ и $U_i(X)$.  Согласование рядов называется иногда сращиванием. Это дало название всему методу сращиваемых асимптотических разложений.


Возьмем $n$-частичную функцию внешнего разложения и представим ее через внутреннюю переменную $X$ 
$$
\tilde{y}_n (x, \varepsilon) = \sum \limits_{i=0}^{n} \varepsilon^i y_i(x) = \sum \limits_{i=0}^{n} \varepsilon^i y_i(\varepsilon X)
$$
Функции $y_i(\varepsilon X)$ можно разложить в ряд при малых $\varepsilon$ и фиксированных значениях $X$:
$$
y_i(\varepsilon X) = \sum \limits_{j=0}^{k} \varepsilon^j y_{ij}(X).
$$
Подставляя второе разложение в первое получим
$$
\tilde{y}_n (x, \varepsilon) =  \sum \limits_{i=0}^{n} \sum \limits_{j=0}^{k} \varepsilon^{i+j} y_{ij}(X).
$$

Аналогично, возьмем $k$-частичную функцию внeвнутреннего разложения и представим ее через внешнюю переменную $х$ 
$$
\tilde{Y}_k (X, \varepsilon) = \sum \limits_{j=0}^{k} \varepsilon^j Y_i(X) = \sum \limits_{j=0}^{k} \varepsilon^j Y_j(x/\varepsilon)
$$
Функции $Y_j(x/\varepsilon)$ можно разложить в ряд при малых $\varepsilon$ и фиксированных значениях $x$:
$$
Y_j(x/\varepsilon) = \sum \limits_{j=0}^{n} \varepsilon^i Y_{ji}(x).
$$
Подставляя второе разложение в первое получим
$$
\tilde{Y}_k (X, \varepsilon) = \sum \limits_{j=0}^{k} \sum \limits_{j=0}^{n} \varepsilon^{i+j} Y_{ji}(x).
$$

Так как существует область в которой $\tilde{y}_n (x, \varepsilon)$ и $\tilde{Y}_k (X, \varepsilon)$ должны совпадать
при $\varepsilon \rightarrow 0$, поэтому
$$
\sum \limits_{i=0}^{n} \sum \limits_{j=0}^{k} \varepsilon^{i+j} y_{ij}(x/\varepsilon) = \sum \limits_{j=0}^{k} \sum \limits_{j=0}^{n} \varepsilon^{i+j} Y_{ji}(x).
$$

Приравнивая в этом выражении коэффицинты при одинаковых степенях $\varepsilon$, получим недостающие условия для внутренней и внешней краевых задач. Эти задачи могут разделяться (мы рассмотрим такие примеры), а могут и не разделяться, и тогда внутреннюю и внешнюю задачи надо будет решать совместно (методом итераций).

Зная решения внутренней и внешней задач можно получить функцию, которая равномерно аппроксимирует искомую функцию $y(x, \varepsilon)$ во всей области $\Omega \cup \partial \Omega$:
$$
y(x, \varepsilon) = \tilde{y}_n (x, \varepsilon) + \tilde{Y}_k (x/\varepsilon, \varepsilon) - \sum \limits_{i=0}^{n} \sum \limits_{j=0}^{k} \varepsilon^{i+j} y_{ij}(x/\varepsilon)
$$

Такая функция называется составным ассимптотическим разложением. 

На этом изложение общей схемы метода сращиваемых ассимптотических разложений закончено. Рассмотрим конкретные примеры.


## Иллюстрация метода сращивания для известной функции

Для иллюстрации метода сращивания рядов рассмотрим функцию 
$$
y(x,\varepsilon) = \frac{e^x}{x+\varepsilon}, \quad  x \in [0,1], \quad \varepsilon >0
$$
При малых $\varepsilon$ эта функция может быть представлена рядом
$$
y(x,\varepsilon) = \frac{e^x}{x} \left( 1 - \frac{\varepsilon}{x}+ \frac{\varepsilon^2}{x^2} - \frac{\varepsilon^3}{x^3}+...\right) = \frac{e^x}{x} \sum \limits_{j=0}^{\infty} (-1)^j \left( \frac{\varepsilon}{x} \right)^j
$$

Однако, этот ряд хорошо приближает исходную функцию везде за исключением окрестности 0. Это внешнее разложение.
Область $\Gamma$ - окрестность нуля.

Введем внутреннюю переменную $X = x/\varepsilon$:
$$
y(x,\varepsilon) = Y(X,\varepsilon ) = \frac{e^{\varepsilon X}}{\varepsilon(1+X)}
$$
Разложим эту функцию в ряд по $\varepsilon$:
$$
Y(X,\varepsilon) = \frac{1}{\varepsilon (1+X)} \left( 1 + \varepsilon X + \varepsilon^2 \frac{X^2}{2}+...\right) 
= \sum \limits_{j=0}^{\infty} \frac{1}{\varepsilon (1+X)}  \varepsilon^j \frac{X^j}{j!}
$$
Однако, этот ряд хорошо приближает исходную функцию в области погранслоя. Это внутреннее разложение.

Проверим, будут ли внешнее и внутренее разложение сращиваться. В соответствии с методом сращиваемых ассимпототических разложений:

1. Перепишем внешнее разложение (возьмем N первых членов) через внутреннюю переменную и разложим в ряд по $\varepsilon$ при фиксированной X.

$$
y(x,\varepsilon) = \sum \limits_{j=0}^{N} (-1)^j \frac{ e^{\varepsilon X}}{\varepsilon X^{j+1}} =
\sum \limits_{j=0}^{N} (-1)^j \frac{ \sum \limits_{i=0}^{K} (\varepsilon X)^i/i!}{\varepsilon X^{j+1}} =
\sum \limits_{j=0}^{N}   \sum \limits_{i=0}^{K}  (-1)^j \frac{  X^{i-j-1}}{i!} \varepsilon^{i-1}
$$

2. Перепишем внутреннее разложение (возьмем K членов) через внешнюю переменную и разложим в ряд по $\varepsilon$ при фиксированной x

$$
Y(X,\varepsilon)  
= \sum \limits_{i=0}^{K} \frac{1}{\varepsilon+x}   \frac{x^i }{i!}
= \sum \limits_{i=0}^{K}   \frac{x^{i-1}}{i!} \sum \limits_{j=0}^{N}  (-1)^j \left( \frac{\varepsilon}{x} \right)^j
= \sum \limits_{i=0}^{K} \sum \limits_{j=0}^{N} (-1)^{j} \frac{x^{i-j-1}}{i!}  \varepsilon^j
$$

Полученные выше двойные разложения совпадают.

Составное ассипмтотическое решение
$$
y_{comb}(x,\varepsilon) = \sum \limits_{j=0}^{N} (-1)^j \frac{ \varepsilon^j e^{х}}{ x^{j+1}} +
\sum \limits_{i=0}^{K} \frac{1}{\varepsilon+x}   \frac{x^i }{i!}
- \sum \limits_{i=0}^{N} \sum \limits_{j=0}^{N} (-1)^{j} \frac{x^{i-j-1}}{i!}  \varepsilon^j
$$





```python
import numpy as np
import matplotlib.pyplot as plt
N = 51 # число точек
eps = 0.1
x = np.linspace(0.1*eps, 1, N)

#x = np.linspace(0, 1, N)
y = np.exp(x)/(x+eps)
z=eps/x



y_out = np.exp(x)/x # разложение до N=0
y_in = 1/(eps+x) # разложение до K=0
y_ass = 1/x

y_out = np.exp(x)/x*(1-z) # разложение до N=1
y_in = (1+x)/(eps+x) # разложение до K=1
y_ass = (1-eps/x)*(1+1/x)

#y_out = np.exp(x)/x*(1-z+z**2) # разложение до N=2
#y_in = (1+x+ x**2/2)/(eps+x) # разложение до K=2
#y_ass = (1/x+1+x/2)*(1-eps/x+(eps/x)**2) 



y_av= y_out+y_in - y_ass                   
                            
plt.ylim(0,10)
#plt.xlim(0.4,0.6)
plt.plot(x, y, linestyle='-', color='k')
plt.plot(x, y_in, linestyle='-', color='b')
plt.plot(x, y_out, linestyle='-', color='r')
plt.plot(x, y_ass, linestyle='-.', color='y')    
plt.plot(x, y_av, linestyle='-', color='y')                             

plt.xlabel('Ось x')
plt.ylabel('y(x)')

plt.show()
```

###### Рассмотрим простое дифференциальное уравнение

\begin{equation}
\varepsilon \frac{ d^2y}{dx^2} + \frac{dy}{dx} +y  = 0, 
\end{equation}

со следующими граничными условиями 

$$ 
y(0) = \alpha,  y(1) = \beta.
$$

Эта краевая задача имеет простое аналитическое решение

$$
y(x; \varepsilon) = \frac{(\alpha e^{s_2}-\beta)e^{s_1 x}+(\beta -\alpha e^{s_1})e^{s_2 x}}{e^{s_2} - e^{s_1}},
$$
где 
$$
s_{1,2} = \frac{-1 \pm \sqrt{1-4\varepsilon}}{2 \varepsilon}
$$
Составим простую программу на питоне и нарисуем решение, например, для $\alpha = 0.5$, $\beta = 0.5$.

```python

N = 10000 # число точек
a = 0.5 
b= 0.5
eps = 0.1
1
x = np.linspace(0, 1, N)
s1 = (-1 + np.sqrt(1-4*eps))/2/eps
s2 = (-1 - np.sqrt(1-4*eps))/2/eps
y=((a*np.exp(s2)-b)*np.exp(s1*x)+(b-a*np.exp(s1))*np.exp(s2*x))/(np.exp(s2)-np.exp(s1))       

plt.plot(x, y, linestyle='-', color='k')
plt.xlabel('Ось x')
plt.ylabel('y(x)')
plt.show()
```

Из рисунка ясно, что при малых значениях $\varepsilon$ решение имеет очень большую производную при x$\rightarrow$0, поэтому первым членом в уравнении пренебречь нельзя.

Попробуем получить решение данной задачи при малых значениях $\varepsilon$  методом сращиваемых асимптотических разложений. 

Получим внешнее решение (то есть решение вне окрестности x=0). 

##### Замечание 1.  Строго говоря, мы должны математически определить, что областью погранслоя является окрестность точки $x=0$, но для данной задачи мы это видим из точного решения, поэтому эту часть задачи пока опустим.

Ищем решение в виде ряда
$$
y(x; \varepsilon) = y_0(x)+ \varepsilon y_1(x) + \varepsilon^2 y_2(x) +.... 
$$
##### Замечание 2.  Для некоторых задач такое разложение не проходит и надо искать калибровочную последовательность

Подставляя в уравнение имеем следующую цепочку уравнений:
$$
\frac{dy_0}{dx}+y_0 =0 \\
\frac{dy_1}{dx}+y_1 = - \frac{d^2y_0}{dx^2} \\
\frac{dy_2}{dx}+y_2 = - \frac{d^2y_1}{dx^2} \\
...
$$
Полученные уравнения являются уравнениями первого порядка, поэтому мы можем найти решения, которые будут удовлетворять
граничному условию $y(1)=\beta$. Возьмем для $y_0, y_1, y_2$ следующие краевые условия: $y_0 (1)= \beta$, $y_1 (1)= 0$,  $y_2 (1)= 0$. Решения соответствующих краевых задач легко найти:
$$
y_0 = \beta e^{1-x} , y_1 =\beta(1-x)e^{1-x}, y_2 =\beta e^{1-x}\left(-\frac{x^2}{2}+3x-\frac{5}{2}\right)
$$

Таким образом, решение внешней задачи с точностью до второго порядка по $\varepsilon$ имеет вид:
$$
y(x) = y_0 + \varepsilon y_1 + \varepsilon^2 y_2 = \beta e^{1-x} 
\left[ 1 + \varepsilon (1-x) + \varepsilon^2 \left(-\frac{x^2}{2}+3x-\frac{5}{2}\right)  \right]
$$
(Важно заметить, что в данном случае внешнее решение определяется независимо от внутреннего. Для задач с частными производными это не всегда так.)



Построим внутреннее решение. Сделаем замену переменной $X = x/\varepsilon$ и положим $y(x; \varepsilon) = Y(x; \varepsilon)$.

Тогда исходное уравнение и граничные условия запишутся в виде
\begin{equation}
 \frac{ d^2Y}{dX^2} +  \frac{dY}{dX}  + \varepsilon Y =  0,   Y(0) =0, Y(1/\varepsilon)=1.
\end{equation}

Оставляем только внутреннее граничное условие и ищем решение в виде $Y(X; \varepsilon) =Y_0(X) + \varepsilon Y_1(X) + \varepsilon^2 Y_2(X) +....$.

Подставляя в уравнение имеем цепочку
\begin{eqnarray}
\begin{cases}
    \frac{d^2 Y_0}{dX^2}+\frac{d Y_0}{dX} =0 \\
    Y_0 (0)= \alpha
  \end{cases}
\end{eqnarray}
\begin{eqnarray}
\begin{cases}
   \frac{d^2 Y_1}{dX^2}+\frac{d Y_1}{dX} =- Y_0 \\
    Y_1 (0)= 0
  \end{cases}
\end{eqnarray}
\begin{eqnarray}
\begin{cases}
   \frac{d^2 Y_2}{dX^2}+\frac{d Y_2}{dX} =- Y_1 \\
    Y_2 (0)= 0
  \end{cases}
\end{eqnarray}


Решая соответствующие задачи, получим
$$
Y_0(X) = \alpha -A_0 (1-e^{-X}),
$$
$$
Y_1(X) = A_1(1-e^{-X}) - \left[ \alpha -A_0 (1+e^{-X}) \right] X,
$$
$$
Y_2(X) = .....
$$

Таким образом, решение внутренней задачи с точностью до первого порядка по  $\varepsilon$  имеет вид:
$$
Y(X) = \alpha -A_0 (1-e^{-X})+\varepsilon 
\left[ A_1(1-e^{-X}) - \left[ \alpha -A_0 (1+e^{-X}) \right] X \right]
$$


Теперь нужно срастить внешнее и внутреннее решения. 

## Сращивание одночленных решений

Возьмем из внешнего и внутреннего разложений по одному члену:
$$
y(x) = \beta e^{1-x},
Y(X) = \alpha -A_0 (1-e^{-X})
$$
В этом случае можно использовать метод сращивания Прандтля (не работающий в следующих приближениях):
$$
lim_{x \rightarrow 0} y(x) =  lim_{X \rightarrow \infty} Y(X).
$$
Отсюда получаем
$$
\beta e = \alpha - A_0   =>    A_0 = \alpha - \beta e 
$$

Таким образом, в этом приближении решение может быть записано

\begin{eqnarray}
y(x) =
  \begin{cases}
    \beta e^{1-x}      & \quad \text{для фиксированного} \quad  x >0, \\
    \beta e +(\alpha - \beta e) e^{-x/\varepsilon} & \quad \text{для фиксированного} \quad X = x/\varepsilon,
   \end{cases}
\end{eqnarray}

В записи выше непонятно, какой формулой пользоваться для конкретных значений $x$ и $\varepsilon$.
Чтобы избежать этой сложности составим решение из комбинации
$$
y_{\text{сост}}(x) = y(x)+Y(x/eps) - \beta e = \left[ \beta e^{1-x} +  \beta e +(\alpha - \beta e) e^{-x/\varepsilon} \right] - \beta e
=  \beta e^{1-x} +  (\alpha - \beta e) e^{-x/\varepsilon} 
$$

Такое решение называется составным.

Построим эти решения для значений $\alpha$, $\beta$ и $\varepsilon$, как на предыдущем графике.


```python
x = np.linspace(0.,10*eps, N)
y=((a*np.exp(s2)-b)*np.exp(s1*x)+(b-a*np.exp(s1))*np.exp(s2*x))/(np.exp(s2)-np.exp(s1))       
y_out=b*np.exp(1-x)
A1= b*np.exp(1)
A0 = a-A1
y_in=A1+A0*np.exp(-x/eps)
y_av= y_out+y_in - A1
y_ass= np.full(N, A1)
plt.plot(x, y, linestyle='-', color='k')
plt.plot(x, y_in, linestyle='-', color='r')
plt.plot(x, y_out, linestyle='-', color='b')
plt.plot(x, y_av, linestyle='-', color='y')
plt.plot(x, y_ass, linestyle='-.', color='y')
plt.xlabel('Ось x')
plt.ylabel('Функция y')
plt.show()
```

### Сравним составное одночленное разложение с точным

```python
#plt.figure(figsize=(15, 12))
plt.plot(x/eps, abs(y_av-y)/y, linestyle='-', color='r', label='Разность значений составного решения первого порядка и аналитического решения')
plt.xlabel('Ось x')
plt.ylabel('Функция f')
plt.legend()
plt.grid()
plt.show()
```

## Сращивание двучленных разложений

Возьмем из внешнего и внутреннего разложений по два члена.

$$
\tilde{y}_1(x) = y_0 + \varepsilon y_1  = \beta e^{1-x} 
\left[ 1 + \varepsilon (1-x) \right]
$$
$$
\tilde{Y}_1(X) = \alpha -A_0 (1-e^{-X})+\varepsilon 
\left[ A_1(1-e^{-X}) - \left[ \alpha -A_0 (1+e^{-X}) \right] X \right]
$$

Найдем общую часть внутреннего и внешних решений.

Представим внешнее решение через внутреннюю переменную $X$ и разложим его в ряд по $\varepsilon$ при фиксированном $X$.
Оставив два члена разложения имеем
$$
\tilde{y}_1(x)  = \beta e^{1-\varepsilon X} 
\left[ 1 + \varepsilon (1-\varepsilon X)  \right] = \beta e (1-\varepsilon X) 
\left[ 1 + \varepsilon (1-\varepsilon X)  \right] =
=  \beta e ( 1-\varepsilon X + \varepsilon)   = \beta e (1-x+ \varepsilon).
$$

Представим внутренее решение через внешнюю переменную $x$ и разложим его в ряд по $\varepsilon$ при фиксированном $x$.
Оставив два члена разложения имеем
$$
\tilde{Y}_1(X) = \alpha -A_0 (1-e^{-x/\varepsilon })+\varepsilon 
\left[ A_1(1-e^{-x/\varepsilon}) - \left[ \alpha -A_0 (1+e^{-x/\varepsilon }) \right] x/\varepsilon \right]
 = \alpha -A_0 +\varepsilon 
\left[ A_1 - \left(  \alpha -A_0 \right)  x/\varepsilon \right] = (\alpha -A_0) (1-x) +  \varepsilon A_1.
$$

В сращиваемых ассимптотических разложений полагается, что существует область, где 
$\tilde{y}_1(x) = \tilde{Y}_1(x/\varepsilon)$:
$$
 \beta e (1-x+ \varepsilon) = (\alpha -A_0) (1-x) +  \varepsilon A_1  \quad  =>  \quad   A_0 = \alpha - \beta e, A_1 = \beta e
$$

Составным ассимптотическим решением в этом случае будет являться функция
$$
y_{\text{сост}}(x) = \tilde{y}_1(x)+\tilde{Y}_1(x/eps) - \beta e (1-x+ \varepsilon)
$$

## Численное решение изначальной краевой задачи

```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt
from scipy.integrate import solve_bvp

def full_sol(x, a=0.5, b=0.5, eps=1e-5 ):
    s1 = (-1 + np.sqrt(1-4*eps))/2/eps
    s2 = (-1 - np.sqrt(1-4*eps))/2/eps
    y=((a*np.exp(s2)-b)*np.exp(s1*x)+(b-a*np.exp(s1))*np.exp(s2*x))/(np.exp(s2)-np.exp(s1))
    return y


def fun(x, y, eps=1e-5):
    return np.vstack((y[1], -(y[1] + y[0]) / eps))

def bc(ya, yb):
    return np.array([ya[0] - 0.5, yb[0] - 0.5])

x = np.linspace(0, 1, 1000)
y_a = np.zeros((2, x.size))
y_a[0] = 0.5
y_a[1] = 0.5
res_a = solve_bvp(fun, bc, x, y_a, max_nodes=1000)

x_plot = np.linspace(0, 1, 1000)
y_plot_a = res_a.sol(x_plot)[0]
y_true = full_sol(x_plot)
plt.figure(figsize=(15, 12))
plt.plot(x_plot, y_true, label='Аналитическое решение', color='g')
plt.plot(x_plot, y_plot_a, label='Численное решение', color='b', linestyle=':')
plt.title('epsilon=0.01')
plt.legend()
plt.xlabel('Ось x')
plt.ylabel('Функция y')
plt.grid()
plt.show()
```

##  ПРИМЕР 2

###### Рассмотрим дифференциальное уравнение

\begin{equation}
\varepsilon^2 \ddot{u} -q(x)u=f(x),  \quad  0 \le x \le 1,  q(x) > const >0. 
\end{equation}

со следующими краевыми условиями 

$$ 
u(0; \varepsilon) = R_0,  u(1; \varepsilon) = R_1.
$$


Разложим в ряд по $\varepsilon$: $~u=u_0+\varepsilon:2 u_2+ \varepsilon^4 u_4+...$  

(подумайте, почему только четные степени)

Подставляя ряд в уравнение и приравнивая члены по одинаковых степенях $\varepsilon$  имеем цепочку
$$
u_0 = f(x)/q(x), u_2 = \ddot{u}_0/q(x),  u_4 = \ddot{u}_2/q(x),....
$$
Таким образом, все члены ряда можно определить, но найденное решение не будет удовлетворять ни одному из краевых условий.
Говорят, что возникла невязка в граничных условиях.  Невязка состоит в том, что при $\varepsilon \rightarrow 0$ в окрестности точек $x=0$ и $x=1$ решение будет от краевых значений $R_0$ и $R_1$ до значений $\approx u_0(0)$ и $\approx u_0(1)$, соответственно.

Рассмотрим точку $x=0$.  Будем искать внутреннюю переменную в виде $X = x/\varepsilon^{\alpha}$, ($0 \lt \alpha $). 
Введем функцию $U(X; \varepsilon) = u(\varepsilon^{\alpha} x; \varepsilon)$. Исходное дифференциальное уравнение перепишется в виде:
$$
\varepsilon^{2-2\alpha} \frac{\partial^2 U}{\partial X^2} -q(X\varepsilon^{\alpha})U=f(X\varepsilon^{\alpha})
$$

Рассмотрим какие члены этого уравнения являются главными при $\varepsilon \rightarrow 0$ для разных $\alpha$:

1.  При $\alpha < 1$ первый член этого уравнения оказывается малым (при  $\varepsilon \rightarrow 0$) (как и в случае внешнего разложения). Этим членом мы уже пренебрегали во внешнем разложении. Поэтому при  $\alpha < 1$ ничего нового по сравнению с внешним разложением мы получить не сможем.

2.  При $\alpha > 1$ первый член становится главным. Пренебрегая остальными членами - получим: 
$$
\frac{d^2 U}{d X^2} =0.
$$
Решение этого уравнения линейная функция, которая неспособна описать плавный переход от внешнего решения к R_0.

3. При $\alpha = 1$ все три члена уравнения становятся одного порядка:
$$
\frac{d^2 U}{d X^2} -q(X\varepsilon^{\alpha})U=f(X\varepsilon^{\alpha}).
$$
Решение этого уравнения способно обеспечить плавный переход. Поэтому внутренняя переменная в этом случае $X= x/\varepsilon$.
Так как уравнение неоднородное, что ищем решение в виде частного решения неоднородного уравнения - за него можно взять внешнее решение, так как оно удовлетворяет уравнению, и общего решения однородного уравнения:
$$
U (X, \varepsilon) = Z(X,\varepsilon) +u(X/\varepsilon, \varepsilon).
$$
Разложение для второго члена мы получили выше. Первый член является решением однородного уравнения
$$
\frac{d^2 Z}{d X^2} -q(X\varepsilon^{\alpha})Z=0.
$$
Ищем его решение в виде ряда
$$
Z(X,\varepsilon) = \sum\limits_{k=0}^{\infty} z_k \varepsilon^k
$$
Разложим функцию $q(x)$  в ряд при малых x
$$
q(\varepsilon X) = \sum\limits_{k=0}^{\infty} q_k \varepsilon^k X^k
$$
Подставляя разложения для $Z$ и $q$ и приравнивая члены при равных степенях $\varepsilon$  получим цепочку

$$
\ddot{z}_0 - q_0 z_0 = 0, 
$$
$$
\ddot{z}_1 - q_0 z_1 = q_1 z_0 X,
$$
$$
\ddot{z}_2 - q_0 z_2 = q_1 z_0 X - q_2 X^2 , 
$$
$$
.....
$$
Граничные условия для этих ОДУ ставятся из следующих соображений:

1.  Функция $Z$ не должна давать (заметный) вклад во внешних областях: $Z(X,\varepsilon) = 0$ при $X \rightarrow \infty$, 
а значит и $z_k(X,\varepsilon) = 0$ при $X \rightarrow \infty$.

2. Для того, чтобы ликвидировать невязку при x=0, внутреннее решение должно удовлетворять граничному условию 
$U (0, \varepsilon) = R_0$.  $=>$  
$$
Z(0,\varepsilon)= R_0 - u(0, \varepsilon).
$$
Используя разложения в ряд для $Z$ и $u$ последнее равенство можно удовлетворить следующим образом
$$
z_0(0) = R_0 - u_0(0),  z_1(0) = 0, z_{2k}(0) = - u_{2k}(0),z_{2k+1}(0) = 0,   k=1,2,3... 
$$


Таким образом, краевые задачи для функций $z_k$ поставлены.
Решения для первых трех функций $u_k$:
$$
z_0(X) = (R_0 - u_0(0))e^{-X\sqrt{q}_0},
$$
$$
z_1(X) = \frac{q_1}{4q_0} (R_0 - u_0(0))(X+X^2\sqrt{q}_0)e^{-X\sqrt{q}_0},
$$
$$
z_2(X) = ...
$$

Таким образом, мы построили ряд  $\tilde{u} (x,\varepsilon) =  \sum\limits_{i=0}^{\infty} (z_i(x/\varepsilon)+u_i (x)) \varepsilon^i$,
который удовлетворяет уравнению и граничному условию при x=0.

Для ликвидации невязки в точке $x=1$ действуем аналогично. Введем внутреннюю переменную $X_1 = (1-x)/\varepsilon$ и обозначим 
$U_1(X_1, \varepsilon)$. Исходное уравнение примет тот же вид, что и для внутреннего разложения в окрестности точки $x=0$
$$
\frac{d^2 U_1}{d X_1^2} -q(X\varepsilon^{\alpha})U_1=f(X_1\varepsilon^{\alpha}).
$$
Далее полностью аналогично предыдущему.
В результате находим разложение в ряд для функции $Z_1(X_1, \varepsilon)$.

Решение удовлетворяющее обоим граничным условиям:
$$
u (x,\varepsilon) =  \sum\limits_{i=0}^{\infty} (u_i(x)+z_i(x/\varepsilon)+z_{1,i}((1-x)/\varepsilon)) \varepsilon^i.
$$
